```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open('Pictures/lena.jpeg')
im_array = np.array(image)
```

```python
plt.imshow(im_array)
plt.title('Image originale')
plt.axis('off')
plt.show()
```

```python
def cadre_noir(im, ep):
    im_cadre = im.copy()
    im_cadre[:ep, :, :] = 0          
    im_cadre[-ep:, :, :] = 0         
    im_cadre[:, :ep, :] = 0          
    im_cadre[:, -ep:, :] = 0         
    return im_cadre
    
```

```python
epaisseur = 5
image_cadre = cadre_noir(im_array, epaisseur)
```

```python
plt.imshow(image_cadre)
plt.title(f'Image avec cadre noir de {epaisseur} pixels')
plt.axis('off')
plt.show()
```
